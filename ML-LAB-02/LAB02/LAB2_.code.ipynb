```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np


data = {
    'Employee_ID': [101, 102, 103, 104, 101, 105, 106, 107],
    'Age': [25, 30, np.nan, 45, 25, 35, 200, 28], 
    'Salary': [50000, 60000, 55000, np.nan, 50000, 65000, 40000, 48000],
    'Department': ['IT', 'HR', 'IT', 'Marketing', 'IT', 'HR', 'IT', 'Marketing'],
    'Left_Company': ['No', 'Yes', 'No', 'No', 'No', 'Yes', 'No', 'Yes']
}
df = pd.DataFrame(data)


sns.set_theme(style="whitegrid")

plt.figure(figsize=(6, 4))
sns.histplot(df['Age'].dropna(), kde=True, bins=10, color='skyblue')
plt.title('Age Distribution Histogram')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()


numeric_df = df.select_dtypes(include=[np.number])

plt.figure(figsize=(6, 4))
sns.heatmap(numeric_df.corr(), annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title('Correlation Heatmap')
plt.show()
```
