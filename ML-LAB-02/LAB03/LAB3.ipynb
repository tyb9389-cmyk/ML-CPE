```python
import pandas as pd
import numpy as np

# สร้างข้อมูลจำลองที่มีปัญหา (เพื่อส่งให้โค้ดคลีนนิ่งบรรทัดถัดๆ ไปของคุณทำงานได้)
data = {
    'Employee_ID': [101, 102, 103, 104, 101, 105, 106, 107],
    'Age': [25, 30, np.nan, 45, 25, 35, 200, 28], 
    'Salary': [50000, 60000, 55000, np.nan, 50000, 65000, 40000, 48000],
    'Department': ['IT', 'HR', 'IT', 'Marketing', 'IT', 'HR', 'IT', 'Marketing'],
    'Left_Company': ['No', 'Yes', 'No', 'No', 'No', 'Yes', 'No', 'Yes']
}
df = pd.DataFrame(data)
```

```python
print("--- Before Cleaning ---")
print(df)
print("-" * 50)


df['Age'] = df['Age'].fillna(df['Age'].median())

df['Salary'] = df['Salary'].fillna(df['Salary'].mean())


df = df.drop_duplicates().reset_index(drop=True)


median_age = df['Age'].median()
df.loc[df['Age'] > 100, 'Age'] = median_age


df['Employee_ID'] = df['Employee_ID'].astype(str)


print("--- After Cleaning ---")
print(df)
print("\n--- Remaining Missing Values ---")
print(df.isnull().sum())
```
