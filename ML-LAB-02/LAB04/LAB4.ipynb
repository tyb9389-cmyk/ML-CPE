```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder


data = {
    'Employee_ID': [101, 102, 103, 104, 105, 106, 107],
    'Age': [25, 30, 31, 45, 35, 31, 28],
    'Salary': [50000, 60000, 55000, 53000, 65000, 40000, 48000],
    'Department': ['IT', 'HR', 'IT', 'Marketing', 'HR', 'IT', 'Marketing'], 
    'Education_Level': ['Bachelor', 'Master', 'PhD', 'Bachelor', 'Master', 'Bachelor', 'Master'], 
    'Left_Company': ['No', 'Yes', 'No', 'No', 'Yes', 'No', 'Yes'] 
}

df = pd.DataFrame(data)

print("--- ข้อมูลตั้งต้นก่อนทำ Encoding ---")
print(df)
print("\n" + "="*60 + "\n")



df_label = df.copy()

le = LabelEncoder()
df_label['Left_Company_Encoded'] = le.fit_transform(df_label['Left_Company'])


education_mapping = {'Bachelor': 0, 'Master': 1, 'PhD': 2}
df_label['Education_Encoded'] = df_label['Education_Level'].map(education_mapping)

print("--- [1] ผลลัพธ์หลังทำ Label Encoding ---")
print(df_label[['Education_Level', 'Education_Encoded', 'Left_Company', 'Left_Company_Encoded']])
print("\n" + "="*60 + "\n")





df_onehot = pd.get_dummies(df, columns=['Department'], prefix='Dept', dtype=int)

print("--- [2] ผลลัพธ์หลังทำ One-Hot Encoding (คอลัมน์ Department) ---")
print(df_onehot.filter(like='Dept_'))
print("\n" + "="*60 + "\n")



df_final = df.copy()


df_final['Left_Company'] = le.fit_transform(df_final['Left_Company'])


df_final['Education_Level'] = df_final['Education_Level'].map(education_mapping)


df_final = pd.get_dummies(df_final, columns=['Department'], prefix='Dept', dtype=int)


df_final = df_final.drop(columns=['Employee_ID'])

print("--- [3] Dataset สมบูรณ์ที่ผ่านการ Feature Engineering พร้อมนำเข้า ML ---")
print(df_final.head())
```
