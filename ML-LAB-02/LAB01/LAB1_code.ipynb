```python
import pandas as pd
import numpy as np


data = {
    'Employee_ID': [101, 102, 103, 104, 101, 105, 106, 107],
    'Age': [25, 30, np.nan, 45, 25, 35, 200, 28],  # มี missing และ outlier (อายุ 200)
    'Salary': [50000, 60000, 55000, np.nan, 50000, 65000, 40000, 48000], # มี missing
    'Department': ['IT', 'HR', 'IT', 'Marketing', 'IT', 'HR', 'IT', 'Marketing'],
    'Left_Company': ['No', 'Yes', 'No', 'No', 'No', 'Yes', 'No', 'Yes'] # Target Class
}
df = pd.DataFrame(data)


print("--- Data Shape ---")
print(df.shape) 
print("\n")


print("--- Data Types ---")
print(df.dtypes)
print("\n")


print("--- Summary Statistics ---")
print(df.describe(include='all')) 
print("\n")


print("--- Missing Values ---")
print(df.isnull().sum())
print("\n")


print("--- Duplicate Records ---")
print(f"Total duplicate rows: {df.duplicated().sum()}")
print("\n")


print("--- Class Distribution (Left_Company) ---")
print(df['Left_Company'].value_counts())
```
